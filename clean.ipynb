### Merging Vehicle Count Percentages

#### Everyday Roadway Vehicles (Class F1 - F7)
ACT_AVG_MC_PERC – The percentage of Motorcycles, class F1, based on the average weekday
distribution of vehicle classes from the most recent vehicle classification count for the segment.
                                                                                 
ACT_AVG_CAR_PERC – The percentage of Passenger Vehicles, class F2, based on the average
weekday distribution of vehicle classes from the most recent vehicle classification count for the
segment.

ACT_AVG_LT_PERC – The percentage of Light Duty Trucks, class F3, based on the average
weekday distribution of vehicle classes from the most recent vehicle classification count for the
segment.

ACT_AVG_SU_PERC – Single Unit Truck and Bus percentage. The percentage of the total
vehicles in classes F4 through F7 based on the average weekday distribution of vehicle classes
from the most recent vehicle classification count for the segment.

#### Everyday Roadway Vehicles (Class F8 - F13)
** Needs Subtraction **

ACT_AVG_TRUCK_PERC – Truck percentage or Heavy Vehicle Percentage. The percentage of
the total vehicles in classes F4 through F13 based on the average weekday distribution of
vehicle classes from the most recent vehicle classification count for the segment.

```python
trimmed_data['ACT_AVG_HEAVY_PERC'] = trimmed_data['ACT_AVG_TRUCK_PERC'] - trimmed_data['ACT_AVG_SU_PERC']
trimmed_data = trimmed_data.dropna(subset=['AADT'])
```

```python
# Save the modified DataFrame back to a CSV
trimmed_data.to_csv('./Count_Statistics_2021_Trimmed_CSV.csv', index=False)
```

## Merging years to try to replace lost data

```python
# Load your datasets
df1 = pd.read_csv("Count_Statistics_2022_Trimmed_CSV.csv")  # Latest year (keep all rows)
df2 = pd.read_csv("Count_Statistics_2021_Trimmed_CSV.csv")  # Previous year (fill missing values)

# Set index to 'COUNT_STATS_ID' (or any key column) to ensure correct matching
df1.set_index("COUNT_STATS_ID", inplace=True)
df2.set_index("COUNT_STATS_ID", inplace=True)

# Fill missing values in df1 using df2 (only fills NaNs, doesn't overwrite existing data)
df1.update(df2)

# Reset index if needed
df1.reset_index(inplace=True)

# Save the cleaned file
df1.to_csv("data_2022_filled.csv", index=False)
```
